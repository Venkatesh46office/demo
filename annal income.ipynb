```python
income=pd.DataFrame()
income['Annual_Income']=an['Annual Income (k$)']
income['frq']=income.Annual_Income.map(income.Annual_Income.value_counts())
income.drop_duplicates(inplace=True)
print(income)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,8))
xrange=range(len(income))
print(xrange)
plt.bar(xrange ,list(income['frq'].values))
plt.xticks(xrange,list(income['Annual_Income'].values))
plt.show()           
```
